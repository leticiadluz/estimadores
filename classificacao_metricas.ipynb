## Avaliao de Classificao

- Modelos de classificao so fundamentais no aprendizado de mquina, sendo amplamente utilizados para categorizar dados em grupos especficos. Exemplos clssicos incluem a identificao de emails como spam ou no-spam, e a classificao de imagens para detectar se contm um gato, um cachorro ou outro objeto. Diferente dos modelos de regresso, que estimam valores contnuos, os modelos de classificao organizam os dados em categorias bem definidas. A preciso com que esses modelos realizam tais classificaes  crucial e, portanto, sua avaliao  uma etapa chave no desenvolvimento de solues eficazes de aprendizado de mquina.

## Mtricas de Avaliao

### 1 Acurcia 
- **A acurcia  uma mtrica intuitiva na avaliao de modelos de classificao, representando a proporo de previses corretas em relao ao total.** Por exemplo, se um modelo preditivo  aplicado para determinar se clientes de uma operadora de telefonia cancelaro seus servios (churn) e acerta as previses em 70 dos 100 casos testados, sua acurcia  de 70%. Matematicamente, a acurcia pode ser expressa pela seguinte frmula:



- **Onde:**
    - Verdadeiros Positivos (VP) so os casos em que o modelo corretamente prediz a classe positiva.
    - Verdadeiros Negativos (VN) so os casos em que o modelo corretamente prediz a classe negativa.
    - Total de Observaes  a soma de todas as previses feitas pelo modelo, incluindo verdadeiros positivos, verdadeiros negativos, falsos positivos e falsos negativos.
- **Apesar de sua simplicidade, a acurcia pode ser uma mtrica enganosa, especialmente em contextos de dados desbalanceados.** Por exemplo, considere um modelo desenvolvido para identificar fraudes em transaes com cartes de crdito, onde apenas 0,5% das transaes so fraudulentas. Um modelo (ruim) que classifique todas as transaes como no fraudulentas atingiria uma acurcia de 99,5%, apesar de ser completamente ineficaz em sua tarefa principal: detectar fraudes. Esse exemplo ilustra por que confiar unicamente na acurcia para avaliar classificadores pode levar a concluses equivocadas.

- **Cenrio:**
    - Total de transaes: 5000
    - Fraudes (0,5%): 25 transaes so fraudes.
    - Transaes legtimas: 4975 transaes.
- Vamos supor que o modelo faa as seguintes previses:
    - VP (Verdadeiros Positivos): **O modelo incorretamente identifica 0 fraudes.**
    - VN (Verdadeiros Negativos): **O modelo corretamente classifica 4975 transaes legtimas.**
- Neste caso, teramos:

- **Nesse cenrio, o modelo atinge a acurcia de 99,5%, mas isso ocorre porque ele classificou corretamente todas as transaes legtimas, porm falhou completamente em identificar qualquer transao fraudulenta.**

### 2 Matriz de Confuso

- A matriz de confuso  uma ferramenta crucial na avaliao de modelos de classificao, estruturada como uma tabela que reflete as frequncias de classificao para cada classe prevista pelo modelo. **Ela se compe de quatro segmentos principais: Verdadeiros Positivos (VP), Falsos Positivos (FP), Verdadeiros Negativos (VN) e Falsos Negativos (FN).**
- Consideremos, por exemplo, a previso de churn em uma base de clientes. Existem quatro possveis desfechos que um modelo pode prever:
    - O modelo prediz que o cliente vai cancelar o servio (churn) e ele realmente cancela (VP).
    - O modelo prediz que o cliente vai cancelar o servio, mas ele no cancela (FP).
    - O modelo prediz que o cliente no vai cancelar o servio e ele realmente no cancela (VN).
    - O modelo prediz que o cliente no vai cancelar o servio, mas ele cancela (FN).
    
    


- **Perceba que a diagonal principal de uma matriz de confuso  formada pelos quadrantes de Verdadeiros Positivos (VP) e Verdadeiros Negativos (VN). Estes valores so particularmente significativos porque representam as previses corretas do modelo**
---
- A partir da matriz de confuso podemos calcular outras mtricas:
- Vamos utilizar novamente o exemplo do desenvolvimento de um modelo de machine learning para deteco de fraudes:
- **Taxa de Verdadeiros Positivos (TPR- True Positive Rate) ou Sensibilidade ou Recall (1FNR):**
    
- Proporo dos verdadeiros positivos em relao ao total de casos positivos reais. 
- Representa a proporo de positivos reais (fraudes, no caso) que foram corretamente identificados como positivos pelo modelo.
---
- **Taxa de Falsos Positivos (FPR - False Positive Rate), (1TNR):**
    
- Proporo dos falsos positivos em relao ao total de casos negativos reais.
- Representa a proporo de negativos reais (transaes legtimas) que foram incorretamente classificados como positivos (fraudes) pelo modelo.
---
- **Taxa de Falsos Negativos (FNR - False Negative Rate), (1TPR):**
    
- Proporo de falsos negativos em relao ao total de casos positivos reais.
- Representa a proporo de positivos reais (fraudes) que foram incorretamente classificados como negativos (transaes legtimas) pelo modelo. Ou seja,  a taxa de fraudes que o modelo deixou de detectar.
---
- **Taxa de Verdadeiros Negativos (TNR- True Negative Rare) ou Especificidade, (1FPR):**
    
-  Proporo de verdadeiros negativos em relao ao total de casos negativos reais.
- Representa a proporo de negativos reais (transaes legtimas) que foram corretamente identificados como negativos pelo modelo. Ou seja,  a taxa de transaes legtimas corretamente classificadas.
---
- Conforme o limite de deciso (representado pela linha preta)  modificado,  possvel aumentar o nmero de positivos detectados (verdadeiros positivos), ou seja, aumentar a taxa de deteco de fraudes. Isso ocorre porque, ao mover o limite para a esquerda, mais exemplos que estavam sendo classificados como negativos (transaes legtimas) passam a ser classificados como positivos (fraudes). No entanto, essa mudana tambm pode aumentar a quantidade de falsos positivos, j que algumas transaes legtimas podem ser incorretamente classificadas como fraudulentas.


- **O limite de deciso que escolhemos sempre depende das prioridades do problema que estamos lidando. Se o objetivo  minimizar os falsos negativos, ou seja, garantir que o maior nmero possvel de positivos (como fraudes ou doenas) seja identificado corretamente, ento podemos mover o limite para a esquerda, o que aumenta a sensibilidade (verdadeiros positivos), mas tambm pode aumentar os falsos positivos.**
- **Por outro lado, se o foco  minimizar os falsos positivos, ou seja, evitar classificar incorretamente negativos como positivos (por exemplo, em investimentos em campanhas publicitrias, evitando gastar recursos com usurios que no sero convertidos), movemos o limite de deciso para a direita. Isso diminui o nmero de falsos positivos, mas pode aumentar o nmero de falsos negativos.**

### 3 Preciso (precision)
 -  **A preciso  a mtrica que avalia a proporo das previses verdadeiras positivas em relao ao total de previses positivas feitas por um modelo**. Matematicamente, ela  calculada como a razo entre os Verdadeiros Positivos (VP) e a soma dos Verdadeiros Positivos (VP) e Falsos Positivos (FP). **Ou seja, a preciso mede a proporo de previses positivas que so efetivamente corretas.** A frmula da preciso  a seguinte:
 
 
 
- **Onde:**:
    - VP (Verdadeiros Positivos): So os casos onde o modelo corretamente identifica um resultado positivo.
    - FP (Falsos Positivos): So as situaes onde o modelo prediz erroneamente um resultado positivo.


- **Exemplo de uso:**
    - Imagine que uma empresa est investindo em uma campanha de marketing digital para atrair novos clientes para um produto especfico. O objetivo  converter o maior nmero possvel de pessoas que realmente tm interesse em comprar o produto. Nesse cenrio, uma alta preciso  essencial para garantir que o investimento na campanha esteja sendo utilizado de forma eficiente, focando o oramento em pessoas que realmente tm maior probabilidade de comprar o produto (VP), minimizando o desperdcio com leads que no convertem (FP).
--- 

### 4 Recall (revocao)
- O recall, tambm conhecido como sensibilidade, revocao ou taxa de verdadeiro positivo, ** a mtrica que indica a proporo de casos positivos reais corretamente identificados pelo modelo em relao ao total de casos positivos reais.** O foco do recall  na cobertura completa dos casos positivos existentes. Isso significa que o recall  crucial em situaes onde no se pode deixar passar nenhum caso positivo. A frmula do recall  expressa da seguinte forma:



- **Onde:**
    - VP (Verdadeiros Positivos): So os casos em que o modelo corretamente identifica um resultado positivo.
    - FN (Falsos Negativos): So os casos em que o modelo no consegue identificar um resultado positivo que de fato existe.
    
- **Exemplo de uso:**
    - Por exemplo, ao trabalhar com um modelo de diagnstico mdico, o objetivo  identificar todas as pessoas doentes. Neste contexto, a capacidade do modelo de detectar todos os indivduos com a doena (todos os verdadeiros positivos)  essencial, pois no detectar uma pessoa doente(FN) pode ter consequncias srias. Por isso, um alto recall  prioritrio, mesmo que isso signifique aceitar mais falsos positivos, isto , diagnosticar como doente algum que na verdade est saudvel.

### 5 F1-Score
- **O F1-Score  uma mtrica valiosa para avaliar modelos de classificao, pois equilibra a preciso e o recall**, integrando-os em um nico indicador por meio da mdia harmnica. A mdia harmnica  usada no clculo do F1 score para combinar preciso e revocao, que so razes com denominadores diferentes. A mdia harmnica  ideal para esse tipo de combinao porque penaliza valores baixos de forma mais significativa do que a mdia aritmtica, garantindo que tanto preciso quanto revocao sejam altas para obter um F1 score elevado. Assim, ela trata as duas mtricas de maneira justa, equilibrando seus impactos.
- Se usssemos a mdia aritmtica, isso no refletiria corretamente a situao em que uma das mtricas  muito baixa, j que ela no penaliza discrepncias to fortemente quanto a mdia harmnica.  **Como resultado, o classificador s obter uma alta pontuao F1 se tanto o recall quanto a preciso forem altos.**
- A frmula para calcular o F1-Score :



- Ela  calculada dividindo-se o nmero de elementos pelo somatrio dos inversos de cada valor no conjunto. A frmula geral da mdia harmnica (Mh) para um conjunto de n elementos :

- Para o exemplo da mdia harmnica com o conjunto de valores (2, 4, 5, 6), a frmula aplicada :

### 6 - Trade-off entre Preciso e Recall

- Vamos treinar um modelo apenas para exemplicar o trade-off entre preciso e recall.
- **Infelizmente,  medida que aumentamos uma das mtricas, geralmente h uma perda na outra.**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.linear_model import LogisticRegression
from feature_engine.selection import SmartCorrelatedSelection as SCS
from sklearn.feature_selection import SequentialFeatureSelector as SFS
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score, auc
```

```python
df_diabetes = pd.read_csv('diabetes.csv')
df_diabetes.head()
```

```python
df_diabetes.info()
```

```python
X = df_diabetes.drop('Outcome', axis = 1)
y = df_diabetes['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 25)
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
numerical_transformer = Pipeline(steps=[
    ('scaler', RobustScaler())
])

preprocessor = ColumnTransformer(
    transformers =[],
    remainder = numerical_transformer
)

base_model = LogisticRegression(class_weight='balanced', random_state=42)

sfs = SFS(estimator=base_model,
                        n_features_to_select='auto',
                        direction='backward',
                        scoring='recall_macro',
                        cv=5)

scs = SCS(method = 'pearson',
         threshold = 0.8,
         selection_method= 'variance')

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('feature_corr', scs ),
    ('feature_selection', sfs)
])

pipeline.fit(X_train, y_train)
```

```python
numeric_features = X_train.columns.tolist()
feature_selection_step = pipeline.named_steps['feature_selection']
print(f'Melhor subconjunto de variveis: {np.array(numeric_features)[feature_selection_step.get_support()]}')
```

```python
X_train_clean = X_train.drop(['BloodPressure','SkinThickness', 'Insulin'], axis=1)
X_train_clean
```

```python
numerical_transformer = Pipeline(steps=[
    ('scaler', RobustScaler())
])

preprocessor = ColumnTransformer(
    transformers = [],
    remainder = numerical_transformer
)

base_model = LogisticRegression(class_weight='balanced')

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', base_model)
])

SF = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

param_grid = {
    'classifier__C': [0.05, 0.1, 0.15, 0.2, 0.5, 1, 10],
}

grid_search_lr = GridSearchCV(pipeline, param_grid, cv=SF, scoring='recall_macro', return_train_score=True)
grid_search_lr.fit(X_train_clean, y_train)                                
```

```python
print(f'Melhores Parmetros: {grid_search_lr.best_params_}')
print(f'Melhor score de validao cruzada: {grid_search_lr.best_score_:.3f}')
```

```python
y_scores = grid_search_lr.predict_proba(X_train_clean)[:, 1]
```

### 6.1 Curva PR(Preciso/Recall):

```python
precisions, recalls, thresholds = precision_recall_curve(y_train, y_scores)
pr_auc = auc(recalls, precisions)
```

```python
plt.plot(recalls, precisions, linewidth=2, label=f'PR-AUC = {pr_auc:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend(loc='upper right')
plt.title('Precision/Recall curve')
plt.grid(True)
plt.show()
```

- Conforme podemos verificar no grfico, h um ponto em que o recall atinge um valor maior que 0.8, e a preciso ainda est acima de 0.6, mostrando um equilbrio razovel entre essas mtricas. No entanto, ao buscar valores de preciso mais elevados, notamos que o recall diminui drasticamente, caindo para valores extremamente baixos. 
- Vamos encontrar o ponto timo em que mantemos um recall maior que 0.8 e a preciso ainda est acima de 0.6

```python
for recall, precision, threshold in zip(recalls, precisions, thresholds):
    if recall >= 0.8 and precision >= 0.6:
        plt.plot(recall, precision, 'ro')
        plt.text(recall, precision, f'({recall:.2f}, {precision:.2f})', color='red')
        print(f'Threshold: {threshold:.2f}, Recall: {recall:.2f}, Precision: {precision:.2f}')
        break

plt.plot(recalls, precisions, linewidth=2, label=f'PR-AUC = {pr_auc:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend(loc='upper right')
plt.title('Precision/Recall curve')
plt.grid(True)
plt.show()
```

### 7 Curva ROC 
- A curva ROC (receiver operating characteristic)  outra ferramenta comum usada com classificadores binrios. Ela  muito semelhante  curva de preciso/recall, mas em vez de plotar a preciso contra a recall, a curva ROC plota a taxa de verdadeiros positivos (recall) contra a taxa de falsos positivos (FPR). A FPR (tambm chamada de queda ou fall-out)  a razo de instncias negativas que so incorretamente classificadas como positivas. 
- Ela exibe a capacidade de um modelo de classificao em distinguir entre as classes em vrios limiares de deciso. Esta curva  construda plotando a Taxa de Verdadeiros Positivos (TPR) contra a Razo de Falsos Positivos (FPR) em diferentes pontos de corte de probabilidade. 



- Mais uma vez, h um trade-off: quanto maior a revocao (TPR), mais falsos positivos (FPR) o classificador produz.  Conforme aumentamos a revocao, o modelo tende a classificar mais casos como positivos para garantir que est encontrando todos os verdadeiros positivos. No entanto, isso acaba aumentando tambm o nmero de falsos positivos. 
-  A linha pontilhada representa a curva ROC de um classificador puramente aleatrio; um bom classificador se mantm o mais distante possvel dessa linha (em direo ao canto superior esquerdo).
- Uma maneira de comparar classificadores  medir a rea sob a curva (AUC). Um classificador perfeito ter uma ROC AUC igual a 1, enquanto um classificador puramente aleatrio ter uma ROC AUC igual a 0,5.
- Conforme j mencionado anteriormente,  se modificarmos o limite de deciso para esquerda,  possvel aumentar o nmero de positivos detectados (verdadeiros positivos), mas tambm aumentamos o nmero de falsos positivos. 



- Por outro lado, se ajustarmos o limite para direita mais verdadeiros negativos sero detectados (verdadeiros negativos) mas aumentaremos o nmero de falsos negativos. 

- Vamos traar uma curva ROC para o nosso modelo:

```python
fpr, tpr, thresholds = roc_curve(y_train, y_scores)
calculate_auc = roc_auc_score(y_train, y_scores)
```

```python
plt.plot(fpr, tpr, color='#4169E1', label='AUC = %0.3f' % calculate_auc)
plt.plot([0, 1], [0, 1], color='#696969', linestyle='--')
plt.xlim([-0.015, 1.01])
plt.ylim([-0.011, 1.05])
plt.xlabel('Taxa de Falsos Positivos (1-Especificidade)')
plt.ylabel('Taxa de Verdadeiros Positivos (Recall)')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.show()
```

- Vamos verificar o Threshold para o ponto temos um recall prximo de 0.9 e uma taxa de falsos positivos prximo de 0.4:

```python
for fpr_val, tpr_val, threshold in zip(fpr, tpr, thresholds):
    if fpr_val <= 0.4 and tpr_val >= 0.9:
        plt.plot(fpr_val, tpr_val, 'ro') 
        plt.text(fpr_val, tpr_val, f'({fpr_val:.2f}, {tpr_val:.2f})', color='red')
        print(f'Threshold: {threshold:.2f}, FPR: {fpr_val:.2f}, TPR: {tpr_val:.2f}')
        break

plt.plot(fpr, tpr, color='#4169E1', label='AUC = %0.3f' % calculate_auc)
plt.plot([0, 1], [0, 1], color='#696969', linestyle='--')
plt.xlim([-0.015, 1.01])
plt.ylim([-0.011, 1.05])
plt.xlabel('Taxa de Falsos Positivos (1-Especificidade)')
plt.ylabel('Taxa de Verdadeiros Positivos (Recall)')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.show()
```

### 8 Curva ROC Ou Curva PR
- **A ROC-AUC  til em conjuntos de dados equilibrados**, pois d uma viso geral do desempenho do modelo, considerando tanto os Verdadeiros Positivos (TPR) quanto os Falsos Positivos (FPR). Contudo, em conjuntos desbalanceados (quando uma classe domina a outra), a ROC-AUC pode ser enganosa.
- J a PR-AUC foca em preciso e recall, o que ** mais til quando a classe positiva  rara(dados desbalanceados)**, ou quando os falsos positivos tm um grande impacto no negcio. Por exemplo, em deteco de spam, voc quer garantir que a maioria dos e-mails que so classificados como spam realmente sejam spam, logo PR-AUC pode ser melhor.
---
- Vamos utilizar um exemplo para ilustrar porque a Curva PR  melhor para dados desbalanceados:
    - Total de exemplos: 10.000
    - Classe Positiva (minoritria): 500 (pessoas com a doena)
    - Classe Negativa (majoritria): 9.500(pessoas sem a doena)
- Agora, vamos supor que temos um classificador com os seguintes resultados:
- Verdadeiros Positivos (TP): 450(o modelo previu corretamente 450 pessoas com a doena)
- Falsos Negativos (FN):50 (o modelo no detectou 50 pessoas com a doena)
- Verdadeiros Negativos (TN): 9.200 (o modelo previu corretamente 9.200 pessoas sem a doena)
- Falsos Positivos (FP): 300 (o modelo disse que 300 pessoas sem a doena tinham a doena)
---
- Clculo da ROC-AUC:
    - **Taxa de Verdadeiros Positivos (TPR ou Recall):** TP/(TP+FN) = 450/(450+50) = 0.9
    - **Taxa de Falsos Positivos (FPR):** FP/(FP/TN) = 300/(300+9.200)  = 0.0316
       - Agora, veja que a FPR  muito baixa (0.0316). Isso ocorre porque temos muitos Verdadeiros Negativos (TN = 9.200) em comparao com os Falsos Positivos (FP = 300), o que 'diminui o valor da FPR, mesmo que haja 300 erros de classificao.
        - Ento, a ROC-AUC pode parecer boa, j que a FPR  pequena, mesmo que o nmero de Falsos Positivos seja grande. O modelo parece estar funcionando bem na ROC-AUC, pois a maioria das pessoas est na classe negativa (9.500).

```python
TP = 450  
FN = 50  
TN = 9200 
FP = 300

y_true = np.array([1]*TP + [0]*FP + [0]*TN + [1]*FN)
y_scores = np.array([1]*TP + [1]*FP + [0]*TN + [0]*FN)
```

```python
fpr, tpr, _ = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, color='blue', label=f'ROC AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
plt.title('Curva ROC - Exemplo 2')
plt.legend(loc='lower right')
plt.show()
```

- Clculo da PR-AUC:
    - **Taxa de Verdadeiros Positivos (TPR ou Recall):** TP/(TP+FN) = 450/(450+50) = 0.9
    - **Preciso (Precision):** TP/(TP+FP): 450(450+300): 0.6
- Aqui, a Preciso (0.6)  muito mais afetada pelos Falsos Positivos (FP = 300), o que reflete o fato de que o modelo est cometendo muitos erros ao prever a classe positiva (pessoas sem a doena sendo classificadas como tendo a doena). O foco na classe positiva faz com que a PR-AUC seja mais baixa em comparao  ROC-AUC, refletindo melhor o impacto dos erros no modelo.

```python
precision, recall, _ = precision_recall_curve(y_true, y_scores)
pr_auc = auc(recall, precision)

plt.plot(recall, precision, label=f'PR AUC = {pr_auc:.2f})')
plt.xlabel('Recall (Taxa de Verdadeiros Positivos)')
plt.ylabel('Preciso')
plt.title('Curva Precision-Recall')
plt.legend(loc='upper right')
plt.show()
```

- Esses grficos mostram a diferena de performance do modelo com base nas duas mtricas, enfatizando como cada uma responde de maneira diferente  classe positiva e ao desbalanceamento de dados.
