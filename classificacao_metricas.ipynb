## Avaliação de Classificação

- Modelos de classificação são fundamentais no aprendizado de máquina, sendo amplamente utilizados para categorizar dados em grupos específicos. Exemplos clássicos incluem a identificação de emails como spam ou não-spam, e a classificação de imagens para detectar se contêm um gato, um cachorro ou outro objeto. Diferente dos modelos de regressão, que estimam valores contínuos, os modelos de classificação organizam os dados em categorias bem definidas. A precisão com que esses modelos realizam tais classificações é crucial e, portanto, sua avaliação é uma etapa chave no desenvolvimento de soluções eficazes de aprendizado de máquina.

## Métricas de Avaliação

### 1 Acurácia 
- **A acurácia é uma métrica intuitiva na avaliação de modelos de classificação, representando a proporção de previsões corretas em relação ao total.** Por exemplo, se um modelo preditivo é aplicado para determinar se clientes de uma operadora de telefonia cancelarão seus serviços (churn) e acerta as previsões em 70 dos 100 casos testados, sua acurácia é de 70%. Matematicamente, a acurácia pode ser expressa pela seguinte fórmula:



- **Onde:**
    - Verdadeiros Positivos (VP) são os casos em que o modelo corretamente prediz a classe positiva.
    - Verdadeiros Negativos (VN) são os casos em que o modelo corretamente prediz a classe negativa.
    - Total de Observações é a soma de todas as previsões feitas pelo modelo, incluindo verdadeiros positivos, verdadeiros negativos, falsos positivos e falsos negativos.
- **Apesar de sua simplicidade, a acurácia pode ser uma métrica enganosa, especialmente em contextos de dados desbalanceados.** Por exemplo, considere um modelo desenvolvido para identificar fraudes em transações com cartões de crédito, onde apenas 0,5% das transações são fraudulentas. Um modelo (ruim) que classifique todas as transações como não fraudulentas atingiria uma acurácia de 99,5%, apesar de ser completamente ineficaz em sua tarefa principal: detectar fraudes. Esse exemplo ilustra por que confiar unicamente na acurácia para avaliar classificadores pode levar a conclusões equivocadas.

- **Cenário:**
    - Total de transações: 5000
    - Fraudes (0,5%): 25 transações são fraudes.
    - Transações legítimas: 4975 transações.
- Vamos supor que o modelo faça as seguintes previsões:
    - VP (Verdadeiros Positivos): **O modelo incorretamente identifica 0 fraudes.**
    - VN (Verdadeiros Negativos): **O modelo corretamente classifica 4975 transações legítimas.**
- Neste caso, teríamos:

- **Nesse cenário, o modelo atinge a acurácia de 99,5%, mas isso ocorre porque ele classificou corretamente todas as transações legítimas, porém falhou completamente em identificar qualquer transação fraudulenta.**
---
### 2 Matriz de Confusão

- A matriz de confusão é uma ferramenta crucial na avaliação de modelos de classificação, estruturada como uma tabela que reflete as frequências de classificação para cada classe prevista pelo modelo. **Ela se compõe de quatro segmentos principais: Verdadeiros Positivos (VP), Falsos Positivos (FP), Verdadeiros Negativos (VN) e Falsos Negativos (FN).**
- Consideremos, por exemplo, a previsão de churn em uma base de clientes. Existem quatro possíveis desfechos que um modelo pode prever:
    - O modelo prediz que o cliente vai cancelar o serviço (churn) e ele realmente cancela (VP).
    - O modelo prediz que o cliente vai cancelar o serviço, mas ele não cancela (FP).
    - O modelo prediz que o cliente não vai cancelar o serviço e ele realmente não cancela (VN).
    - O modelo prediz que o cliente não vai cancelar o serviço, mas ele cancela (FN).
    
    


- **Perceba que a diagonal principal de uma matriz de confusão é formada pelos quadrantes de Verdadeiros Positivos (VP) e Verdadeiros Negativos (VN). Estes valores são particularmente significativos porque representam as previsões corretas do modelo**
---
- A partir da matriz de confusão podemos calcular outras métricas:
- Vamos utilizar novamente o exemplo do desenvolvimento de um modelo de machine learning para detecção de fraudes:
- **Taxa de Verdadeiros Positivos (TPR- True Positive Rate) ou Sensibilidade ou Recall (1−FNR):**
    
- Proporção dos verdadeiros positivos em relação ao total de casos positivos reais. 
- Representa a proporção de positivos reais (fraudes, no caso) que foram corretamente identificados como positivos pelo modelo.
---
- **Taxa de Falsos Positivos (FPR - False Positive Rate), (1−TNR):**
    
- Proporção dos falsos positivos em relação ao total de casos negativos reais.
- Representa a proporção de negativos reais (transações legítimas) que foram incorretamente classificados como positivos (fraudes) pelo modelo.
---
- **Taxa de Falsos Negativos (FNR - False Negative Rate), (1−TPR):**
    
- Proporção de falsos negativos em relação ao total de casos positivos reais.
- Representa a proporção de positivos reais (fraudes) que foram incorretamente classificados como negativos (transações legítimas) pelo modelo. Ou seja, é a taxa de fraudes que o modelo deixou de detectar.
---
- **Taxa de Verdadeiros Negativos (TNR- True Negative Rare) ou Especificidade, (1−FPR):**
    
-  Proporção de verdadeiros negativos em relação ao total de casos negativos reais.
- Representa a proporção de negativos reais (transações legítimas) que foram corretamente identificados como negativos pelo modelo. Ou seja, é a taxa de transações legítimas corretamente classificadas.
---
- Conforme o limite de decisão (representado pela linha preta) é modificado, é possível aumentar o número de positivos detectados (verdadeiros positivos), ou seja, aumentar a taxa de detecção de fraudes. Isso ocorre porque, ao mover o limite para a esquerda, mais exemplos que estavam sendo classificados como negativos (transações legítimas) passam a ser classificados como positivos (fraudes). No entanto, essa mudança também pode aumentar a quantidade de falsos positivos, já que algumas transações legítimas podem ser incorretamente classificadas como fraudulentas.


- **O limite de decisão que escolhemos sempre depende das prioridades do problema que estamos lidando. Se o objetivo é minimizar os falsos negativos, ou seja, garantir que o maior número possível de positivos (como fraudes ou doenças) seja identificado corretamente, então podemos mover o limite para a esquerda, o que aumenta a sensibilidade (verdadeiros positivos), mas também pode aumentar os falsos positivos.**
- **Por outro lado, se o foco é minimizar os falsos positivos, ou seja, evitar classificar incorretamente negativos como positivos (por exemplo, em investimentos em campanhas publicitárias, evitando gastar recursos com usuários que não serão convertidos), movemos o limite de decisão para a direita. Isso diminui o número de falsos positivos, mas pode aumentar o número de falsos negativos.**

---

### 3 Precisão (precision)
 -  **A precisão é a métrica que avalia a proporção das previsões verdadeiras positivas em relação ao total de previsões positivas feitas por um modelo**. Matematicamente, ela é calculada como a razão entre os Verdadeiros Positivos (VP) e a soma dos Verdadeiros Positivos (VP) e Falsos Positivos (FP). **Ou seja, a precisão mede a proporção de previsões positivas que são efetivamente corretas.** A fórmula da precisão é a seguinte:
 
 
 
- **Onde:**:
    - VP (Verdadeiros Positivos): São os casos onde o modelo corretamente identifica um resultado positivo.
    - FP (Falsos Positivos): São as situações onde o modelo prediz erroneamente um resultado positivo.


- **Exemplo de uso:**
    - Imagine que uma empresa está investindo em uma campanha de marketing digital para atrair novos clientes para um produto específico. O objetivo é converter o maior número possível de pessoas que realmente têm interesse em comprar o produto. Nesse cenário, uma alta precisão é essencial para garantir que o investimento na campanha esteja sendo utilizado de forma eficiente, focando o orçamento em pessoas que realmente têm maior probabilidade de comprar o produto (VP), minimizando o desperdício com leads que não convertem (FP).
--- 

### 4 Recall (revocação)
- O recall, também conhecido como sensibilidade, revocação ou taxa de verdadeiro positivo, **é a métrica que indica a proporção de casos positivos reais corretamente identificados pelo modelo em relação ao total de casos positivos reais.** O foco do recall é na cobertura completa dos casos positivos existentes. Isso significa que o recall é crucial em situações onde não se pode deixar passar nenhum caso positivo. A fórmula do recall é expressa da seguinte forma:



- **Onde:**
    - VP (Verdadeiros Positivos): São os casos em que o modelo corretamente identifica um resultado positivo.
    - FN (Falsos Negativos): São os casos em que o modelo não consegue identificar um resultado positivo que de fato existe.
    
- **Exemplo de uso:**
    - Por exemplo, ao trabalhar com um modelo de diagnóstico médico, o objetivo é identificar todas as pessoas doentes. Neste contexto, a capacidade do modelo de detectar todos os indivíduos com a doença (todos os verdadeiros positivos) é essencial, pois não detectar uma pessoa doente(FN) pode ter consequências sérias. Por isso, um alto recall é prioritário, mesmo que isso signifique aceitar mais falsos positivos, isto é, diagnosticar como doente alguém que na verdade está saudável.
---

### 5 F1-Score
- **O F1-Score é uma métrica valiosa para avaliar modelos de classificação, pois equilibra a precisão e o recall**, integrando-os em um único indicador por meio da média harmônica. A média harmônica é usada no cálculo do F1 score para combinar precisão e revocação, que são razões com denominadores diferentes. A média harmônica é ideal para esse tipo de combinação porque penaliza valores baixos de forma mais significativa do que a média aritmética, garantindo que tanto precisão quanto revocação sejam altas para obter um F1 score elevado. Assim, ela trata as duas métricas de maneira justa, equilibrando seus impactos.
- Se usássemos a média aritmética, isso não refletiria corretamente a situação em que uma das métricas é muito baixa, já que ela não penaliza discrepâncias tão fortemente quanto a média harmônica.  **Como resultado, o classificador só obterá uma alta pontuação F1 se tanto o recall quanto a precisão forem altos.**
- A fórmula para calcular o F1-Score é:



- Ela é calculada dividindo-se o número de elementos pelo somatório dos inversos de cada valor no conjunto. A fórmula geral da média harmônica (Mh) para um conjunto de n elementos é:



**Desta forma, para calcular a média harmônica de um conjunto de valores, dividimos o número de elementos(n) pelo denominador que é composto do somatório dos inversos de cada um dos elementos (1/X), lembrando que o inverso de um número é representado pela fração 1 sobre ele mesmo.** 
- Para o exemplo da média harmônica com o conjunto de valores (2, 4, 5, 6), a fórmula aplicada é:
    - Mh= 4 / (1/2) + (1/4) + (1/5) + (1/6)
    - Mh= 4 / (67/60)
    - Mh = 4 x 60/67 -> multiplicando pelo inverso da fração
    - Mh = 240/67 = 3.58

### 6 - Trade-off entre Precisão e Recall

- Vamos treinar um modelo apenas para exemplicar o trade-off entre precisão e recall.
- **Infelizmente, à medida que aumentamos uma das métricas, geralmente há uma perda na outra.**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.linear_model import LogisticRegression
from feature_engine.selection import SmartCorrelatedSelection as SCS
from sklearn.feature_selection import SequentialFeatureSelector as SFS
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score, auc
```

```python
df_diabetes = pd.read_csv('diabetes.csv')
df_diabetes.head()
```

```python
df_diabetes.info()
```

```python
X = df_diabetes.drop('Outcome', axis = 1)
y = df_diabetes['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 25)
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
numerical_transformer = Pipeline(steps=[
    ('scaler', RobustScaler())
])

preprocessor = ColumnTransformer(
    transformers =[],
    remainder = numerical_transformer
)

base_model = LogisticRegression(class_weight='balanced', random_state=42)

sfs = SFS(estimator=base_model,
                        n_features_to_select='auto',
                        direction='backward',
                        scoring='recall_macro',
                        cv=5)

scs = SCS(method = 'pearson',
         threshold = 0.8,
         selection_method= 'variance')

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('feature_corr', scs ),
    ('feature_selection', sfs)
])

pipeline.fit(X_train, y_train)
```

```python
numeric_features = X_train.columns.tolist()
feature_selection_step = pipeline.named_steps['feature_selection']
print(f'Melhor subconjunto de variáveis: {np.array(numeric_features)[feature_selection_step.get_support()]}')
```

```python
X_train_clean = X_train.drop(['BloodPressure','SkinThickness', 'Insulin'], axis=1)
X_train_clean
```

```python
numerical_transformer = Pipeline(steps=[
    ('scaler', RobustScaler())
])

preprocessor = ColumnTransformer(
    transformers = [],
    remainder = numerical_transformer
)

base_model = LogisticRegression(class_weight='balanced')

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', base_model)
])

SF = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

param_grid = {
    'classifier__C': [0.05, 0.1, 0.15, 0.2, 0.5, 1, 10],
}

grid_search_lr = GridSearchCV(pipeline, param_grid, cv=SF, scoring='recall_macro', return_train_score=True)
grid_search_lr.fit(X_train_clean, y_train)                                
```

```python
print(f'Melhores Parâmetros: {grid_search_lr.best_params_}')
print(f'Melhor score de validação cruzada: {grid_search_lr.best_score_:.3f}')
```

```python
y_scores = grid_search_lr.predict_proba(X_train_clean)[:, 1]
```

### 6.1 Curva PR(Precisão/Recall):

```python
precisions, recalls, thresholds = precision_recall_curve(y_train, y_scores)
pr_auc = auc(recalls, precisions)
```

```python
plt.plot(recalls, precisions, linewidth=2, label=f'PR-AUC = {pr_auc:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend(loc='upper right')
plt.title('Precision/Recall curve')
plt.grid(True)
plt.show()
```

- Conforme podemos verificar no gráfico, há um ponto em que o recall atinge um valor maior que 0.8, e a precisão ainda está acima de 0.6, mostrando um equilíbrio razoável entre essas métricas. No entanto, ao buscar valores de precisão mais elevados, notamos que o recall diminui drasticamente, caindo para valores extremamente baixos. 
- Vamos encontrar o ponto ótimo em que mantemos um recall maior que 0.8 e a precisão ainda está acima de 0.6

```python
for recall, precision, threshold in zip(recalls, precisions, thresholds):
    if recall >= 0.8 and precision >= 0.6:
        plt.plot(recall, precision, 'ro')
        plt.text(recall, precision, f'({recall:.2f}, {precision:.2f})', color='red')
        print(f'Threshold: {threshold:.2f}, Recall: {recall:.2f}, Precision: {precision:.2f}')
        break

plt.plot(recalls, precisions, linewidth=2, label=f'PR-AUC = {pr_auc:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend(loc='upper right')
plt.title('Precision/Recall curve')
plt.grid(True)
plt.show()
```

### 7 Curva ROC 
- A curva ROC (receiver operating characteristic) é outra ferramenta comum usada com classificadores binários. Ela é muito semelhante à curva de precisão/recall, mas em vez de plotar a precisão contra a recall, a curva ROC plota a taxa de verdadeiros positivos (recall) contra a taxa de falsos positivos (FPR). A FPR (também chamada de queda ou fall-out) é a razão de instâncias negativas que são incorretamente classificadas como positivas. 
- Ela exibe a capacidade de um modelo de classificação em distinguir entre as classes em vários limiares de decisão. Esta curva é construída plotando a Taxa de Verdadeiros Positivos (TPR) contra a Razão de Falsos Positivos (FPR) em diferentes pontos de corte de probabilidade. 



- Mais uma vez, há um trade-off: quanto maior a revocação (TPR), mais falsos positivos (FPR) o classificador produz.  Conforme aumentamos a revocação, o modelo tende a classificar mais casos como positivos para garantir que está encontrando todos os verdadeiros positivos. No entanto, isso acaba aumentando também o número de falsos positivos. 
-  A linha pontilhada representa a curva ROC de um classificador puramente aleatório; um bom classificador se mantém o mais distante possível dessa linha (em direção ao canto superior esquerdo).
- Uma maneira de comparar classificadores é medir a área sob a curva (AUC). Um classificador perfeito terá uma ROC AUC igual a 1, enquanto um classificador puramente aleatório terá uma ROC AUC igual a 0,5.
- Conforme já mencionado anteriormente,  se modificarmos o limite de decisão para esquerda, é possível aumentar o número de positivos detectados (verdadeiros positivos), mas também aumentamos o número de falsos positivos. 



- Por outro lado, se ajustarmos o limite para direita mais verdadeiros negativos serão detectados (verdadeiros negativos) mas aumentaremos o número de falsos negativos. 

- Vamos traçar uma curva ROC para o nosso modelo:

```python
fpr, tpr, thresholds = roc_curve(y_train, y_scores)
calculate_auc = roc_auc_score(y_train, y_scores)
```

```python
plt.plot(fpr, tpr, color='#4169E1', label='AUC = %0.3f' % calculate_auc)
plt.plot([0, 1], [0, 1], color='#696969', linestyle='--')
plt.xlim([-0.015, 1.01])
plt.ylim([-0.011, 1.05])
plt.xlabel('Taxa de Falsos Positivos (1-Especificidade)')
plt.ylabel('Taxa de Verdadeiros Positivos (Recall)')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.show()
```

- Vamos verificar o Threshold para o ponto temos um recall próximo de 0.9 e uma taxa de falsos positivos próximo de 0.4:

```python
for fpr_val, tpr_val, threshold in zip(fpr, tpr, thresholds):
    if fpr_val <= 0.4 and tpr_val >= 0.9:
        plt.plot(fpr_val, tpr_val, 'ro') 
        plt.text(fpr_val, tpr_val, f'({fpr_val:.2f}, {tpr_val:.2f})', color='red')
        print(f'Threshold: {threshold:.2f}, FPR: {fpr_val:.2f}, TPR: {tpr_val:.2f}')
        break

plt.plot(fpr, tpr, color='#4169E1', label='AUC = %0.3f' % calculate_auc)
plt.plot([0, 1], [0, 1], color='#696969', linestyle='--')
plt.xlim([-0.015, 1.01])
plt.ylim([-0.011, 1.05])
plt.xlabel('Taxa de Falsos Positivos (1-Especificidade)')
plt.ylabel('Taxa de Verdadeiros Positivos (Recall)')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.show()
```

### 8 Curva ROC Ou Curva PR
- **A ROC-AUC é útil em conjuntos de dados equilibrados**, pois dá uma visão geral do desempenho do modelo, considerando tanto os Verdadeiros Positivos (TPR) quanto os Falsos Positivos (FPR). Contudo, em conjuntos desbalanceados (quando uma classe domina a outra), a ROC-AUC pode ser enganosa.
- Já a PR-AUC foca em precisão e recall, o que **é mais útil quando a classe positiva é rara(dados desbalanceados)**, ou quando os falsos positivos têm um grande impacto no negócio. Por exemplo, em detecção de spam, você quer garantir que a maioria dos e-mails que são classificados como spam realmente sejam spam, logo PR-AUC pode ser melhor.
---
- Vamos utilizar um exemplo para ilustrar porque a Curva PR é melhor para dados desbalanceados:
    - Total de exemplos: 10.000
    - Classe Positiva (minoritária): 500 (pessoas com a doença)
    - Classe Negativa (majoritária): 9.500(pessoas sem a doença)
- Agora, vamos supor que temos um classificador com os seguintes resultados:
- Verdadeiros Positivos (TP): 450(o modelo previu corretamente 450 pessoas com a doença)
- Falsos Negativos (FN):50 (o modelo não detectou 50 pessoas com a doença)
- Verdadeiros Negativos (TN): 9.200 (o modelo previu corretamente 9.200 pessoas sem a doença)
- Falsos Positivos (FP): 300 (o modelo disse que 300 pessoas sem a doença tinham a doença)
---
- Cálculo da ROC-AUC:
    - **Taxa de Verdadeiros Positivos (TPR ou Recall):** TP/(TP+FN) = 450/(450+50) = 0.9
    - **Taxa de Falsos Positivos (FPR):** FP/(FP/TN) = 300/(300+9.200)  = 0.0316
       - Agora, veja que a FPR é muito baixa (0.0316). Isso ocorre porque temos muitos Verdadeiros Negativos (TN = 9.200) em comparação com os Falsos Positivos (FP = 300), o que 'diminui o valor da FPR, mesmo que haja 300 erros de classificação.
        - Então, a ROC-AUC pode parecer boa, já que a FPR é pequena, mesmo que o número de Falsos Positivos seja grande. O modelo parece estar funcionando bem na ROC-AUC, pois a maioria das pessoas está na classe negativa (9.500).

```python
TP = 450  
FN = 50  
TN = 9200 
FP = 300

y_true = np.array([1]*TP + [0]*FP + [0]*TN + [1]*FN)
y_scores = np.array([1]*TP + [1]*FP + [0]*TN + [0]*FN)
```

```python
fpr, tpr, _ = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, color='blue', label=f'ROC AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
plt.title('Curva ROC - Exemplo 2')
plt.legend(loc='lower right')
plt.show()
```

- Cálculo da PR-AUC:
    - **Taxa de Verdadeiros Positivos (TPR ou Recall):** TP/(TP+FN) = 450/(450+50) = 0.9
    - **Precisão (Precision):** TP/(TP+FP): 450(450+300): 0.6
- Aqui, a Precisão (0.6) é muito mais afetada pelos Falsos Positivos (FP = 300), o que reflete o fato de que o modelo está cometendo muitos erros ao prever a classe positiva (pessoas sem a doença sendo classificadas como tendo a doença). O foco na classe positiva faz com que a PR-AUC seja mais baixa em comparação à ROC-AUC, refletindo melhor o impacto dos erros no modelo.

```python
precision, recall, _ = precision_recall_curve(y_true, y_scores)
pr_auc = auc(recall, precision)

plt.plot(recall, precision, label=f'PR AUC = {pr_auc:.2f})')
plt.xlabel('Recall (Taxa de Verdadeiros Positivos)')
plt.ylabel('Precisão')
plt.title('Curva Precision-Recall')
plt.legend(loc='upper right')
plt.show()
```

- Esses gráficos mostram a diferença de performance do modelo com base nas duas métricas, enfatizando como cada uma responde de maneira diferente à classe positiva e ao desbalanceamento de dados.
