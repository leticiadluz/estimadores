# Regresso Linear
- A regresso linear  uma tcnica estatstica utilizada para modelar a relao entre uma varivel dependente (ou resposta) e uma ou mais variveis independentes (ou preditoras). **O objetivo  encontrar a equao de uma linha reta que melhor represente a relao entre essas variveis**. Desta forma, ela pode ser considerada um processo estatstico que procura determinar a equao matemtica mais adequada para definir a relao entre duas variveis.



- Ao realizar uma anlise de regresso linear, o objetivo  encontrar uma equao que explique a relao entre as variveis envolvidas.  **Essa equao, envolve coeficientes  que so ajustados de modo a minimizar a diferena entre os valores previstos pela equao e os valores observados nos dados.**
- Olhando para nossos pontos, podemos tentar explic-los com alguma reta, mas note que nenhuma reta cruza todos os pontos, afinal,  impossvel ter um fenmeno que seja perfeitamente descrito por uma reta. O que a gente busca  algo que se aproxime da relao observada, **a reta escolhida ser a que mais se aproximar dos pontos, a que 'erra' menos.**
- Em uma anlise de regresso linear simples, a equao que descreve a relao entre uma varivel dependente y e uma varivel independente X  geralmente representada como:



- Onde:
 
    - y  a varivel dependente (a que est sendo prevista). 
    - X  a varivel independente (a que est sendo usada para prever y). 
    - 0  o coeficiente linear, intercepto da linha de regresso, o ponto onde a reta cruza o eixo y (o valor de y quando X  zero). 
    - 1  o coeficiente angular, indica a inclinao da reta (quanto y varia para cada unidade de variao em X). 
    -   o termo de erro, representando a variao no explicada na varivel dependente.

##  Estimando os Coeficientes 
- De forma resumida, o processo de estimar os coeficientes de um modelo de regresso linear envolve os seguintes passos:
    - Definir a reta de regresso: Para qualquer i-sima observao, a previso para y  dada por y^= B0 + B1* x1.
    - Clculo dos resduos: A diferena entre o valor observado yi e o valor previsto yi^  chamada de resduo ei =yi  yi^.
    - Soma dos quadrados dos resduos (RSS): A RSS  a soma dos quadrados de todos os resduos:
        - RSS=(y1(0^1^x1))+(y2(0^1^x2))+...+(yn(0^1^xn))
    - Estimar os coeficientes: Os coeficientes 0 ^ e  1^ so estimados de maneira que minimizem a RSS e a abordagem mais comum  pelo mtodo dos mnimos quadrados:
        - **Mtodo dos mnimos Quadrados:** O procedimento de mnimos quadrados ** uma abordagem especfica para realizar esse ajustamento linear**. Ele estabelece que a melhor reta  aquela que minimiza a soma dos quadrados das diferenas entre os valores previstos pela reta e os valores reais observados. **Em outras palavras, procuramos minimizar a soma dos quadrados dos resduos (distncias verticais entre os pontos e a reta ajustada)**.
    


- Onde:
    - yi  o valor observado da varivel dependente para o i-simo ponto nos seus dados.
    - yi^  o valor predito da varivel dependente para o i-simo ponto, obtido a partir da equao da linha de regresso.
 

    - **Distncias na Direo Vertical:** Ao considerar a varivel y como aleatria (ou seja, sujeita a variao), **as distncias que estamos tentando minimizar so medidas na direo vertical (y-y^)**. Em um grfico bidimensional, isso significa que estamos considerando as distncias verticais entre cada ponto de dados e a reta de ajuste.

### Termos: 

- **Resduo**: Quando estamos falando do erro para um ponto especfico nos dados de treinamento (diferena entre o valor observado e o valor previsto pelo modelo para esse ponto), **estamos falando do resduo**, isso  representado como ei = yi - yi^, onde yi  o valor observador e yi^  o valor previsto pelo modelo para o ponto i.



- **Termo do erro**: Por outro lado, quando  estamos falando do conjunto de todos os resduos em um conjunto de dados, ou seja, a diferena entre o valor real e o valor previsto para todos os pontos de dados, considerando o modelo como um todo, estamos falando do **Termo do erro**, representado por ().
- **Funo de Custo (Cost Function):** J quando queremos quantificar a discrepncia entre as previses feitas pelo modelo e os valores reais observados nos dados de treinamento, usamos uma mtrica conhecida como **funo de custo**, ela  usada como uma mtrica para avaliar a qualidade das previses do modelo em relao aos dados de treinamento e guiar o processo de treinamento do modelo. No contexto da regresso, uma funo de custo frequentemente usada em problemas de regresso  o Erro Quadrtico Mdio - MSE, mas pode variar dependendo do problema.
    - <span style="background-color:lightgreen">**O objetivo durante o treinamento do modelo  minimizar essa funo de custo, ou seja, ajustar os parmetros do modelo de forma que a discrepncia entre as previses e os valores reais seja a menor possvel**.</span>

## Regresso Linear Mltipla 
-  A regresso linear mltipla  uma extenso da regresso linear simples, onde temos mltiplas variveis independentes para prever uma varivel dependente. A equao da regresso linear mltipla  dada por:



- Onde:
    - y  a varivel dependente (a que est sendo prevista). 
    - X1, X2, ...,Xp so as variveis independentes (ou caractersticas) que esto sendo usadas para prever y. 
    - 0   o intercepto da linha de regresso, que representa o valor de y quando todas as variveis independentes so zero.
    - 1, 2,...,p so os coeficientes angulares, que indicam a mudana esperada  em y para uma unidade de mudana em cada uma das variveis independentes, mantendo todas as outras variveis constantes.
    -   o termo de erro, representando a variao no explicada na varivel dependente.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

### Como queremos apenas demonstrar o uso da regresso linear, no vamos nos preocupar, neste momento, com o pr-processamento dos dados:
- Vamos remover as colunas categricas para no precisarmos trat-las, apenas para este exemplo.
- Na construo de um modelo, devemos trat-las e no simplesmente remov-las.

```python
#Removendo colunas categricas
df = df.drop(['sex', 'smoker','region'], axis=1)
df.head()
```

## Diviso dos dados

```python
X = df.drop('charges', axis =1)
y = df['charges']
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size = 0.3, random_state = 100)
```

```python
modelo = LinearRegression()
modelo.fit(X_treino, y_treino)
```

```python
y_pred = modelo.predict(X_teste)
```

```python
#Avaliando o modelo
mean_absolute_error(y_teste, y_pred)
```

```python
mean_squared_error(y_teste, y_pred)
```

```python
np.sqrt(mean_squared_error(y_teste, y_pred))
```

## Regresso Polinomial
- Quando os dados tm uma relao no linear, um modelo de regresso linear simples pode no capturar bem essa relao. Por exemplo, se seus dados seguem uma curva quadrtica ou cbica, uma linha reta no ser capaz de ajustar esses dados corretamente.
- A regresso polinomial  uma tcnica que permite ajustar modelos lineares a dados no lineares. Fazemos isso transformando as caractersticas originais em novas caractersticas que incluem potncias dessas caractersticas.
- Exemplo de dados no lineares:

- Claramente, uma linha reta nunca se ajustar corretamente a esses dados. **Ento, podemos usar a classe PolynomialFeatures do Scikit-Learn para transformar nossos dados de treinamento, adicionando o quadrado (polinmio de segundo grau) de cada caracterstica no conjunto de treinamento como uma nova caracterstica.**
-  importante escalonar as caractersticas antes de aplicar a regresso polinomial, pois caractersticas com diferentes escalas podem afetar a performance do modelo.

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

- Instanciao da Classe PolynomialFeatures:
    - degree=2: Especifica o grau do polinmio. Neste caso, estamos criando caractersticas polinomiais at o segundo grau (quadrtico). 
    - Para uma caracterstica x, isso incluir x e x. 
    - Para duas caractersticas x1 e x2, isso incluir x1, x2, x1, x2 e x1x2. 
- Exemplo prtico:

```python
# Para uma nica caracterstica
import numpy as np

X = np.array([[2], [3], [4]])
X 
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
X_poly
```

```python
# Para duas caractersticas
X = np.array([[2, 3], [3, 4], [4, 5]])

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
X_poly
```

- Vantagens da Regresso Polinomial:
    - Permite que modelos lineares capturem relaes complexas entre caractersticas e a varivel alvo.
    - Voc pode ajustar o grau do polinmio para capturar mais complexidade nos dados.
- Desvantagens:
    - Exploso Combinatria: O nmero de caractersticas pode crescer rapidamente, tornando o modelo mais complexo e computacionalmente caro.
    - Overfitting: Modelos com polinmios de alto grau podem ajustar-se muito bem aos dados de treinamento, mas no generalizar bem para novos dados
