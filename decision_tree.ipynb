## rvores de Deciso

- rvores de deciso so um tipo de modelo de aprendizado supervisionado usado tanto para classificao quanto para regresso em machine learning. Elas so chamadas assim porque voc pode visualiz-las como uma estrutura em forma de rvore, onde cada 'n' representa uma deciso baseada em uma pergunta sobre os dados.
- **A rvore de deciso  construda como uma espcie de fluxograma, onde as observaes percorrem uma srie de condies determinadas pelas features do modelo at cair numa deciso final a depender do caminho.**



- Sendo uma tcnica capaz de lidar com a no-linearidade, fcil de codificar, com poucas premissas e, talvez o mais importante, com resultado fcil de interpretar, a rvore de decises  comumente adotada para tarefas como concesses de crdito, diagnsticos mdicos, anlises de efeitos de drogas, anlises de investimentos, etc.

### Estrutura de uma rvore de Deciso:
- **N Raiz:** O ponto de partida da rvore. Contm a pergunta ou condio que tem o maior poder discriminatrio.
- **Ns Internos:** Perguntas ou condies adicionais que ajudam a refinar a deciso.
- **Ramos:** Caminhos entre os ns que representam as respostas s perguntas nos ns.
- **Ns Folha:** Os pontos finais da rvore, onde no h mais perguntas a serem feitas. Em tarefas de classificao, cada n folha corresponde a uma classe.

- **Cada um dos ns possuem exatamente 1 pai e 2 filhos:** 
    - Quando um n no possui um pai, o chamamos de n raiz. 
    - Quando ele no possui filhos, chamamos de n folha


### Treinamento 
- Durante a fase de treinamento, o objetivo principal  construir a rvore de deciso a partir dos dados de entrada. Aqui esto os passos envolvidos:
    - **Seleo de Atributos:** O algoritmo escolhe o melhor atributo dos dados para dividir os dados em subgrupos. Este  frequentemente o ponto mais importante e complexo no treinamento de rvores de deciso.
    - **Critrio de Diviso:** Utiliza-se um critrio para escolher qual atributo dividir os dados. O critrio mai comum:
        - **ndice Gini:** O ndice Gini mede a impureza dos ns. Um n  'puro' (Gini=0) se todos os casos nele contidos pertencerem  mesma classe.
    - **Criao de Ns:** Baseado no melhor atributo e critrio, os dados so divididos em dois ou mais subconjuntos, criando novos ns para cada subconjunto.
    - **Recurso:** Este processo  repetido para cada novo n (a no ser que um critrio de parada seja atingido) at que os dados no possam ser divididos de forma significativa. Os critrios de parada comuns incluem profundidade mxima da rvore, nmero mnimo de amostras em um n, ou um ganho mnimo de informao.
    


- **Navegao pela rvore:** Comea-se no n raiz e examinam-se os atributos dos dados de entrada.
    - **N Raiz:** O n raiz inicia a diviso dos dados pela idade dos indivduos. Isso indica que a idade  um fator significativo para determinar o risco de um ataque cardaco.
---
- **Deciso em Cada N:** Em cada n, os dados so avaliados com base na condio ou pergunta do n. Dependendo da resposta, segue-se para o n filho correspondente.
    - **Primeira Diviso:**
        - **Menor que 18:** Se uma pessoa tem menos de 18 anos, ela  direcionada para um n de deciso que avalia o peso.
        - **Entre 18 e 30 anos:** Pessoas nesta faixa etria so automaticamente classificadas como de baixo risco, sem levar em conta outros fatores.
        - **Maior que 30 anos:** Pessoas acima de 30 anos so direcionadas para uma subrvore que avalia se so fumantes.
---
- **Chegada ao N Folha:** Continua-se este processo at chegar a um n folha. O n folha contm a predio ou resultado final, baseado nas condies que foram atendidas ao longo do caminho desde o n raiz.
    - **Peso:**
       - Peso < 60: Se o peso  menor que 60 kg, o risco de ataque cardaco  considerado baixo.
       - Peso > 60: Se o peso  maior que 60 kg, o risco  considerado alto.
    - **Fumante:**
        - No Fumante: Se a pessoa no fuma, o risco continua sendo baixo.
        - Fumante: Se a pessoa fuma, o risco  considerado alto.
        
   - **Ns Folha: Estes so os resultados finais da rvore, indicando se uma pessoa est em 'Baixo Risco' ou 'Alto Risco' de ter um ataque cardaco.**
---

## Hiperparmetros

- **max_depth:** Profundidade mxima. Esse parmetro define o nvel mximo de profundidade da rvore. O n raiz est no nvel de profundidade 0, enquanto os dois filhos do n raiz esto no nvel de profundidade 1, e assim por diante. Naturalmente, voc pode pensar que aumentar o max_depth far com que o modelo se ajuste melhor aos dados e melhore suas previses. No entanto, isso traz o risco de overfitting, onde o modelo se adapta to bem aos dados de treinamento que performa mal em dados novos ou no vistos. Isso realmente pode levar a overfitting, desta forma a reduo deste hiperparmetro evita o ajuste excessivo.
- **min_samples_split:** Este parmetro determina o nmero mnimo de instncias que um n deve ter antes que uma diviso possa ser considerada. Em outras palavras, ele controla a quantidade mnima de amostras necessrias em um n para que este possa ser dividido em ns filhos. Ajustar o min_samples_split  um exerccio de balanceamento. Valores muito altos podem levar a modelos subajustados, onde a rvore no  profunda o suficiente para capturar as relaes importantes nos dados. Por outro lado, valores muito baixos podem permitir que o modelo crie muitos ns com poucas amostras, resultando em overfitting. 
- **min_samples_leaf:** Este parmetro estabelece o nmero mnimo de instncias que um n deve ter para se qualificar como um n folha. Um n folha  o ponto terminal de uma rvore, onde no so feitas mais divises e  feita uma predio.Valores baixos para min_samples_leaf permitem que os ns folha tenham poucas instncias, com um nmero pequeno de instncias em cada n folha, o modelo pode capturar rudos e detalhes especficos dos dados de treinamento, levando a overfitting. J valores altos para min_samples_leaf exigem que cada n folha tenha um nmero maior de instncias, portanto, aumentar o min_samples_leaf ajuda a simplificar o modelo, pois restringe o nmero de divises que a rvore pode fazer. Isso significa que o modelo  menos propenso a aprender o rudo presente nos dados de treinamento.
- **max_leaf_nodes:** Especifica o nmero mximo de ns folha que uma rvore pode ter. Este parmetro desempenha um papel importante na configurao da complexidade da rvore e, consequentemente, na sua capacidade de generalizao. Limitar o nmero de ns folha ajuda a evitar que a rvore se torne excessivamente complexa e especfica aos dados de treinamento, o que poderia prejudicar seu desempenho em novos dados. Por outro lado, um valor muito restritivo pode impedir que o modelo capture padres importantes nos dados, levando a um desempenho subptimo.


## rvores para problemas de Regresso
- rvores de deciso para regresso so uma adaptao do modelo de rvore de deciso tradicionalmente usado para classificao, permitindo prever valores contnuos em vez de categorias discretas. Essas rvores so conhecidas como rvores de regresso e so aplicadas em diversos problemas onde o objetivo  estimar uma varivel dependente contnua a partir de variveis independentes.
- O principal diferencial na rvore de regresso  que, ao invs de prever uma classe, cada n folha prediz um valor contnuo. Normalmente, esse valor  a mdia (ou mediana) dos valores da varivel dependente das amostras que alcanam esse n.
- Mesmo sendo um modelo de regresso, o objetivo principal de uma rvore de deciso de regresso tambm  separar os dados em grupos ou regies com base nas variveis preditoras.

## Critrio de Diviso
- O critrio para decidir como dividir os dados em ns de deciso numa rvore de regresso  baseado na reduo mxima possvel de uma mtrica de erro. Os critrios mais comuns incluem:
    - Erro quadrtico mdio (MSE): A diviso escolhida  aquela que resulta no maior decrscimo possvel no MSE total das divises resultantes.
    - Erro absoluto mdio (MAE): Alternativamente, algumas rvores de regresso podem usar a reduo no MAE para escolher divises.
---    
### MAE:
- O Erro Mdio Absoluto, conhecido pela sigla MAE,  uma das mtricas mais simples de interpretar em uma anlise de regresso. Ele calcula o erro absoluto, ou seja, desconsidera a direo do erro (se foi acima ou abaixo do valor real). 
- Vamos considerar um exemplo hipottico:
    - Valores reais: 40, 50, 35, 50, 44
    - Valores previstos: 39, 40, 30, 45, 40
    - O MAE ser: 
        - |40 - 39| = 1
        - |50 - 40| = 10
        - |35 - 30| = 5
        - |50 - 45| = 5
        - |- 44 - 40| = 4
     - Agora, calculamos a mdia desses erros absolutos: (1 + 10 + 5 + 5 + 4) / 5 = 25 / 5 = 5
    - Portanto, o Erro Mdio Absoluto (MAE) para esse conjunto de dados  de 5 unidades.
### MSE:
- O Erro Quadrtico Mdio, frequentemente abreviado como MSE. Enquanto o MAE se concentra no erro absoluto, o MSE foca no erro quadrtico. Isso significa que ele penaliza erros maiores de forma mais severa do que erros menores. 

- Vamos agora calcular o MSE usando os mesmos valores reais e previstos que mencionamos anteriormente:
    - Valores reais: 40, 50, 35, 50, 44
    - Valores previstos: 39, 40, 30, 45, 40
    - O MSE ser: 
        - |40 - 39| = 1
        - |50 - 40| = 100
        - |35 - 30| = 25
        - |50 - 45| = 25
        - |- 44 - 40| = 16
        
    - Agora, vamos tirar a mdia desses erros quadrticos:(1 + 100 + 25 + 25 + 16) / 5 = 167 / 5 = 33.4
    - Portanto, o Erro Quadrtico Mdio (MSE) para esses valores  de aproximadamente 33.4.
---
- **Clculo do Value:** Quando os dados so divididos at um certo ponto definido por critrios de parada (como profundidade mxima da rvore, nmero mnimo de amostras em um n, etc.), o n resultante torna-se um n folha. O 'value' deste n folha  calculado como a mdia dos valores da varivel alvo de todas as amostras que chegaram a esse n.
- Por exemplo, se um n folha contm as instncias com os valores alvo [10, 20, 30], o 'value' do n ser a mdia destes, ou seja, (10+20+30)/3 = 20.



### Funo de Custo para rvores de Regresso

- Para rvores de regresso, o MSE (Erro Quadrtico Mdio)  comumente usado como funo de custo. O MSE mede a mdia dos quadrados das diferenas entre os valores observados e os valores preditos pelo modelo. Assim como no caso da classificao, o custo  calculado como a soma ponderada dos MSEs dos ns filhos e o objetivo  minimizar a funo de custo:



- Vamos exemplificar o clculo com os dois primeiros ns filhos:
    - MSE ponderado: (5/15 x 0.022) + 10/15 x 2.306)
    - MSE ponderado: 0.007333 + 1.537333  1.5446667
- Assim, o MSE ponderado aproximado aps a diviso  de aproximadamente 1.545. Este valor representa uma mdia ponderada dos erros dos dois ns filhos, levando em considerao o nmero de amostras em cada um
- Ao minimizar essa funo de custo, estamos buscando a diviso que resulta nos menores erros quadrticos mdios ponderados pelos tamanhos dos ns esquerdo e direito. Isso significa que a diviso  escolhida de forma a reduzir ao mximo o erro de previso nos ns filhos, melhorando assim a qualidade da rvore de deciso.

### Por que evitamos codificao one-hot em rvores de deciso?
- Quando codificamos uma varivel categrica usando one-hot encoding, cada categoria se torna uma nova varivel binria (dummy variable).
- Por exemplo, se tivermos uma varivel com 4 categorias (A, B, C, D), ela se transformar em 4 variveis binrias:
    - A: [1, 0, 0, 0]
    - B: [0, 1, 0, 0]
    - C: [0, 0, 1, 0]
    - D: [0, 0, 0, 1]
- **Impacto na Crescimento da rvore:** No exemplo, acima cada categoria tem uma chance de 25% de aparecer e a medida que aumentamos o nmero de categorias esta proporo diminu mais ainda, desta forma, a proporo de valores 1 em relao aos valores 0  pequena, especialmente quando o nmero de categorias aumenta. Isso significa que a maioria dos valores em cada varivel dummy ser 0. Essa distribuio desigual pode afetar a seleo de splits (divises) na rvore de deciso, pois o algoritmo de diviso (como Gini ou entropia) pode no achar essas variveis muito informativas.
- **Distncia do N Raiz:** Como as variveis dummies so independentes entre si, uma diviso em uma dessas variveis geralmente no resulta em um ganho significativo de pureza. Por isso,  improvvel que o algoritmo escolha essas variveis para dividir prximo ao n raiz. As rvores tendem a crescer em direes que minimizam a impureza, e variveis dummies podem no proporcionar essa minimizao significativa, resultando em splits que ocorrem mais longe do n raiz. **Desta forma, rvores de deciso podem ter dificuldade em encontrar splits significativos em variveis dummies, resultando em splits mais afastados do n raiz.**
    - Divises prximas do n raiz so responsveis por grandes redues na impureza (como Gini ou entropia)
    - Variveis que causam divises perto do n raiz so geralmente mais importantes.
- **Impacto na Impureza:** A impureza em uma rvore de deciso  medida pela heterogeneidade dos dados nos ns. Em uma varivel dummy,  maioria dos valores ser 0, o que pode levar a uma diviso que no reduz suficientemente a impureza. Por exemplo, se temos um n com muitas variveis dummies, a diviso baseada em uma delas no criar ns filhos significativamente mais puros, ois a maioria dos dados continuar sendo 0 em outras variveis. A presena predominante de valores 0 em variveis dummies faz com que as divises resultem em ganhos marginais de pureza, impactando negativamente a eficincia da rvore.

### Esses fatores juntos podem fazer com que a rvore de deciso subestime a importncia de variveis codificadas em one-hot, mesmo que essas variveis sejam relevantes para o modelo e assim a rvore de deciso tende a crescer longe dessas variveis quando usamos a codificao one-hot.
